```python
import pandas as pd
import numpy as np
import matplotlib.pyplot  as plt
from matplotlib.collections import LineCollection
#Linear regression - designate a column
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import ElasticNet
```

```python
f = pd.read_csv('file.csv')
X = f.copy()
```

```python
#filter data to only have high non-null columnsets

#do a ranked order list of the X
cols_by_null_rank = sorted([(col,X[col].count()) for col in X.columns],key=lambda x: x[1])
top_non_nulls = [col[0] for col in cols_by_null_rank  if "Equities__" in col[0] and col[1] > 5261 and '___Date' not in col[0]]
#use only the X's columns that have a high Xd. 
Xd = X[top_non_nulls].dropna()
Xd /= Xd.std(axis=0)
```

```python
outcome = Xd['Equities___VIX']
data = Xd.drop('Equities___VIX',axis=1)
trainX = data.ix[:2500,:]
trainY = outcome.ix[:2500]
testX = data.ix[2500:,:]
testY = outcome.ix[2500:]
```

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols
```

```python
trainXConstant = sm.tools.add_constant(trainX, prepend=False)
results =sm.OLS(trainY,trainXConstant).fit()
print results.summary()
```

```python
print results.rsquared
```

```python
#Calculate VIF - Variance Inflation Factor
from statsmodels.stats.outliers_influence import variance_inflation_factor
VIF = []
exogX = trainX.as_matrix()
for i,col in enumerate(trainX.columns):
    VIF.append((col,variance_inflation_factor(exogX,i)))
```

```python
from statsmodels.stats.outliers_influence import OLSInfluence
influence = OLSInfluence(results)
print influence.cooks_distance
print plt.plot(influence.influence)
```

```python
print influence.summary_table()
```

```python
print [i[1] for i in VIF]
plt.hist([i[1] for i in VIF])
```

```python
fig, ax = plt.subplots(figsize=(12,8))
fig = sm.graphics.influence_plot(results,ax=ax,criterion='cooks')
```
