```python
#filter the nan's in the header
def filter(item):
  try: 
      if np.isnan(item):
#print "removing nan"
          return ""
  except:
      if item:
#print "string-stripping"
          return str(item).strip().replace(" ","-")
      else:
          return ""
      
combined_header = []
for i in range(len(h1)):
    tmp = filter(h1[i])+"_"+filter(h2[i])+"_"+filter(h3[i])+"_"+filter(h4[i])
    combined_header.append(tmp)
```

```python
#function removes commas in number
def num(element):
    try:
        element = float(element.replace(",",""))
        return element
    except:
        return element
```

```python
#Define transformation functions

#define a decorator that allows for nan's
def NanSafe(function):
  def wrapper(*args,**kwargs):
      assert(len(list(args))==1)
      x = float(args[0])
      if not np.isnan(x):
          return function(x)
      else: 
          return x
  return wrapper

@NanSafe
def logNanSafe(x):
  return np.log(x)
@NanSafe
def sqNanSafe(x):
  return np.power(x,2)

def lag(df,column,lag):
  df[column+"_Lag{}".format(lag)] = df[column].shift(-1*lag)

def logCol(df,column):
  df[column+"_Log"] = df[column].apply(lambda x: logNanSafe(x))
def square(df,column):
  df[column+"_Sq"] = df[column].apply(lambda x: sqNanSafe(x))
def difference(df,column,lag):
  df[column+"_DiffLag{}".format(lag)] = df[column] - df[column].shift(-1*lag)

def year_difference(df,column):
        return difference(df,column,365)

def run(df,title):    
    lag(df,title,1)
    logCol(df,title)
    square(df,title)
    difference(df,title,1)
```

```python
#define a plotting function that allows for dates plotting
import matplotlib.pyplot as plt
from matplotlib.dates import YearLocator, MonthLocator, DateFormatter

def plotValues(col):
    years = YearLocator()
    months = MonthLocator()
    yearsFmt = DateFormatter('%d-%y')
    
    dates = col.index
    opens = col.values
    
    fig, ax = plt.subplots()
    ax.plot_date(dates, opens, '-')
    
    
    # format the ticks
    ax.xaxis.set_major_locator(years)
    ax.xaxis.set_major_formatter(yearsFmt)
    ax.xaxis.set_minor_locator(months)
    ax.autoscale_view()
    
    # format the coords message box
    def price(x): return 'LIBOR-OIS Spread: %1.2f'%x
    ax.fmt_xdata = DateFormatter('%Y-%m-%d')
    ax.fmt_ydata = price
    ax.grid(True)
    
    fig.autofmt_xdate()
    plt.show()
```

```python
#plot the LIBOR-OIS Spread - expect to see a huge spike in 2007-2010
OIS = 'Rates___3-month-Libor-vs-3-month-OIS-(CALC)'
plotValues(c['Rates___3-month-Libor-vs-3-month-OIS-(CALC)'])
```

```python
#aggregate to month level - decide on whether we should use last, or average in order to 
def aggregate_to_month(date_obj):
    return (date_obj.year,date_obj.month)
month_c_last = c.groupby(aggregate_to_month).last()
month_c_ave = c.groupby(aggregate_to_month).mean()
plt.plot(month_c_last[OIS],'g')
plt.plot(month_c_ave[OIS],'r')
plt.show()
```

```python
#normalize each variable - Z-score: 
def calc_std(o_df):
    df = o_df.copy(deep=True)
    for col in df.columns:
        df[col] = (df[col] - df[col].mean())/df[col].std()
    return df

month_c_ave_std = calc_std(month_c_ave)
```

```python
#after z-score - for each column, look for how many values have > 1.96 values (outliers)
import scipy.stats as st

def list_outliers(df,p_val):
    for col in df.columns:
        print col, (df[col] > st.norm.ppf(p_val)).sum()

list_outliers(month_c_ave_std,.95)
```

```python
#aggregate by feature categories in order to allow color by in the network - took ~3 
def column_type(string):
    if "Rates" in string: 
        return "Rates"
    elif "Commodities" in string:
        return "Commodities"
    elif "Credit" in string:
        return "Credit"
    elif  "Equities" in string:
        return "Equities"
    elif "G10-FX" in string:
        return "G10-FX"
    elif "Municipal-Markets" in string:
        return "Municipal-Markets"


def add_variables(name, grp_df,grp):
    for g in grp.groups:
        if g == name:
            grp_df["Indicator: "+g] = 1
        else: 
            grp_df["Indicator: "+g] = 0
    return grp_df

new_c = c.copy(deep=True)
#replace index with epoch instead of datetime objects
new_c['date_index']=c.index
#convert index to epoch-since 
new_c.index=c.epoch_since
d = new_c.T.groupby(column_type,axis=0)
```

```python
#run add_variables on each of the groups
l = []
for name, group in d: 
    l.append(add_variables(name,group,d))
```

```python
#result has flags for what class the "Feature belongs in" 
result = l[0]
# result.join(l[1])
result = result.append(l[1:])
result.ix[:10,:len(result.columns)-6].to_csv('newDF')
a = result.ix[:,:len(result.columns)-6].T
plt.plot(a['Rates___3-month-Libor-vs-3-month-OIS-(CALC)'])
plt.show()
```
