```python
h1=df.ix[0,:]
h2=df.ix[1,:]
h3=df.ix[2,:]
h4=df.ix[3,:]
```

```python
print len(h1),len(h2),len(h3),len(h4) 
```

```python
def filter(item):
    try: 
        if np.isnan(item):
            print "removing nan"
            return ""
    except:
        if item:
            print "string-stripping"
            return str(item).strip().replace(" ","-")
        else:
            return ""

combined_header = []
for i in range(len(h1)):
    tmp = filter(h1[i])+"_"+filter(h2[i])+"_"+filter(h3[i])+"_"+filter(h4[i])
    combined_header.append(tmp)
```
